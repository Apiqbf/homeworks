```python
#2
import numpy as np
import pandas as pd
a={
    "Аты":["Айжан","Мұрат","Санжар","Алина","Нұржан"],
    "Жасы":[21,19,25,22,23],
    "Баға":[83,74,91,67,88]
}
b=pd.DataFrame(a)
mean=np.mean(b["Баға"])
std=np.std(b["Баға"])
b.loc[len(b)]=["Орташа","NaN",mean]
b.loc[len(b)]=["Ауытқу","NaN",std]
print(b)
```
