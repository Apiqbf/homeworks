```python
#30

def partition_by_sum_parity(s):
    even_set=set()
    odd_set=set()

    for n in s:
        summa=sum(int(d) for d in str(n))
        if summa%2==0:
            even_set.add(n)
        else:
            odd_set.add(n)

    return (even_set,odd_set)

test={23,543,124,455,56,21}

print(partition_by_sum_parity(test))

```

```python
#18
def sort_dict_by_value_sum(d):
    s_dict=sorted(d.items(),
                       key=lambda x:
                       (-sum(x[1]),x[0])
                       )

    return s_dict

dict_by_value={
    "apple": [1, 2, 3],
    "banana": [2, 2, 2],
    "cherry": [5],
    "date": [10, -2],
    "fig": [],
    "grape": [4, 1]
}

print(sort_dict_by_value_sum(dict_by_value))
```

```python
#8
def sort_dict_by_value_length(d):
    sorted_dict=sorted(
        d.items(),
        key=lambda x:
        (len(x[1]),x[0])
    )

    return sorted_dict

func_d={
    "a": "cat",
    "b": "hi",
    "c": "apple",
    "d": "go",
    "e": "sun",
}

result=sort_dict_by_value_length(func_d)

print(result)
```

```python
#4
def filter_sets(sets_list):
    new=[]
    for set in sets_list:
        if len(set)<=3:
            continue

        if any(n<1 for n in set):
            continue

        if sum(n%2==0 for n in set)>=1:
            continue

        new.append(set)

    return new

test_s=[
    {1, 2, 3, 4},
    {0, 2, 4, 6},
    {1, 3, 5},
    {-1, 2, 3, 4},
    {7, 9, 11, 13},
    {2, 4, 6, 8, 10}
]

result=filter_sets(test_s)
print(result)
```
