```python
#2
process_text = lambda s: " ".join(
    w[::-1] for w in s.split()
    if not any(ch.isdigit() for ch in w) and len(w[::-1]) % 2 == 0
)

process_text("hello ab12 test noon")
```

```python
#6
filter_words = lambda s: [
    word for word in s.split()
    if len(word) >= 4
    and word.isalpha()
    and len(set(word)) == len(word)
]

filter_words("suka ya yeball ego vse1")
```

```python
#10
count_words = lambda s: sum(
    1 for word in s.split()
    if len(word) >= 5 and not word[0].isdigit() and any(c.isdigit() for c in word)
)
count_words("abc123 1start middle123 12345 another1word short1 a123")
```

```python
#11
from collections import Counter
def common_unique_chars(s1,s2):
    empty=""
    con=[x for x in s1 if x.isalpha()]+[y for y in s2 if y.isalpha()]
    dic=dict(Counter(con))
    for a,b in dic.items():
        if b>1:
            continue
        empty+=a
    return empty

common_unique_chars("Maulid","Kotakbas")
```

```python
#12
aa=lambda s:[
    word for word in s.split()
    if word[0]==word[-1]
    and word!=word[::-1]
    and len(word)>=3
]

aa("werew цукоац укк")
```

```python
#13
def replace_nth_withchar(s,n,char):
    new=""
    for a in s.split():
        for b in range(len(a)):
            if b==n:
                a[b]==char
            new+=a
    return new

replace_nth_withchar("fsrth sgjher heyhr",3,"hui")

```

```python
#22
def group_by_parity_and_sort(nums):
    even=[]
    odd=[]
    for a in nums:
        if a%2==0:
            even+=[a]
        else:
            odd+=[a]
    return even+odd

group_by_parity_and_sort([1,2,657,5,6,4,18])

```

```python
#26
def remove_duplicates_keep_last(nums):
    result=[]
    for a in reversed(nums):
        if a not in result:
            result+=[a]
    result.reverse()
    return result

remove_duplicates_keep_last([1, 2, 3, 2, 4, 1, 5])
```

```python
#30
def analyze_strings_list(words):
    new=[]
    finished=[]
    for a in words:
        if a==a.isalpha():

        if len(a)%2==0:
            a=a[::-1]
        else:
            a=a.upper()
        new+=[a]

    for b in new:
        if b not in finished:
            finished+=[b]
    return finished

analyze_strings_list(["14f3","qrgrteh","wqegrehteh"])
```

```python
#11
def group_by_length(words):
    result={}
    for w in words:
        l=len(w)
        if l not in result:
            result[l]=[]
        if w not in result:
            result[l]+=[w]
    return result

group_by_length(["lkflkd","dlsfn","sdf","sd","sgg","f","f","a","fwef"])
```

```python
#19
def filter_by_digit_sum(nums):
    result = set()
    for num in nums:
        if num % 2 != 0:
            digit_sum = 0
            for digit in str(abs(num)):
                digit_sum += int(digit)
            if digit_sum % 2 == 0:
                result.add(num)
    return result

filter_by_digit_sum({12, 13, 21, 35, 44})
```

```python
#24
def union_of_filtered_sets(sets_list):
    new=[]
    for b in sets_list:
        for c in b:
            if c<10 or c%2==0:
                continue
            new+=[c]
    print(set(new))

union_of_filtered_sets([{34,3,56,7,8,9,15},{5,6,7,89},{13,3444,50}])
```

```python
#36
from itertools import combinations
def all_subsets_of_size_k(s,k):
    new=[]
    for b in combinations(s,k):
        new+=[set(b)]
    return new


all_subsets_of_size_k({1,2,3,4,5},4)

```

```python
#38
from collections import Counter
def multi_symmetric_difference(sets_list):
    un=[]
    fin=[]
    for b in sets_list:
        for c in b:
            un+=[c]

    d=dict(Counter(un))
    for x,y in d.items():
        if y>1:
            continue
        fin+=[x]

    return f"симметрическая разность={set(fin)}"

multi_symmetric_difference([{3,4,5,6,7,8},{1,2,3,4,5}])
```
