```python
#11
from collections import Counter
def common_unique_chars(s1,s2):
    empty=""
    con=[x for x in s1 if x.isalpha()]+[y for y in s2 if y.isalpha()]
    dic=dict(Counter(con))
    for a,b in dic.items():
        if b>1:
            continue
        empty+=a
    return empty

common_unique_chars("Maulid","Kotakbas")
```

```python
#13
def replace_nth_withchar(s,n,char):
    new=""
    for a in s.split():
        for b in range(len(a)):
            if b==n:
                a[b]==char
            new+=a
    return new

replace_nth_withchar("fsrth sgjher heyhr",3,"hui")

```

```python
#26
def remove_duplicates_keep_last(nums):
    result=[]
    for a in reversed(nums):
        if a not in result:
            result+=[a]
    result.reverse()
    return result

remove_duplicates_keep_last([1, 2, 3, 2, 4, 1, 5])
```

```python
#1
def invert_unique(d):
    new={}
    for x,y in d.items():
        if y not in new:
            new[y]=[]
        if x not in new[y]:
            new[y]+=[x]
    return new

d = {'a': 1, 'b': 2, 'c': 1, 'd': 2, 'e': 3}
print(invert_unique(d))

```

```python
#3
def merge_dicts(d1,d2):
    new=d1.copy()
    for x,y in d2.items():
        if x in new:
            new[x]+=y
        else:
            new[x]=y
    return new

merge_dicts({"ara":1,"hui":3,"eblan":43},{"ara":32,"faah":324,"eblan":1})
```

```python
#9
from collections import Counter
def common_elements_all(sets_list):
    new=[]
    new1=[]
    for a in sets_list:
        for b in a:
            new+=[b]
    for x,y in Counter(new).items():
        if y<len(sets_list):
            continue
        new1+=[x]

    return set(new1)

common_elements_all([{1,2,3,4,5,6},{2,3,4,5},{3,4,6,7}])


```

```python
#11
def group_by_length(words):
    result={}
    for w in words:
        l=len(w)
        if l not in result:
            result[l]=[]
        if w not in result:
            result[l]+=[w]
    return result

group_by_length(["lkflkd","dlsfn","sdf","sd","sgg","f","f","a","fwef"])
```

```python
#19
def filter_by_digit_sum(nums):
    result = set()
    for num in nums:
        if num % 2 != 0:
            digit_sum = 0
            for digit in str(abs(num)):
                digit_sum += int(digit)
            if digit_sum % 2 == 0:
                result.add(num)
    return result

filter_by_digit_sum({12, 13, 21, 35, 44})
```

```python
#32
def pairwise_intersections(sets_list):
    if len(sets_list)<2:
        return []
    else:
        return [sets_list[i] & sets_list[i + 1] for i in range(len(sets_list) - 1)]

print(pairwise_intersections([{1, 2, 3}, {2, 3, 4}, {3, 4, 5}]))
```

```python
#33
filter_by_avg = lambda d: {
    k: v for k, v in d.items()
    if sum(v) / len(v) > (
        sum(sum(lst) for lst in d.values()) /
        sum(len(lst) for lst in d.values())
    )
}

data = {
    "a": [1, 2, 3],
    "b": [10, 20],
    "c": [4, 5, 6]
}

result = filter_by_avg(data)
print(result)

```

```python
#38
from collections import Counter
def multi_symmetric_difference(sets_list):
    un=[]
    fin=[]
    for b in sets_list:
        for c in b:
            un+=[c]

    d=dict(Counter(un))
    for x,y in d.items():
        if y>1:
            continue
        fin+=[x]

    return f"симметрическая разность={set(fin)}"

multi_symmetric_difference([{3,4,5,6,7,8},{1,2,3,4,5}])
```

```python
#Анализ текстовой базы данных студентов

import string

students = [
    {
        "name": "Alice123",
        "grades": [12, 9, 15, 8],
        "comments": ["Good work", "Excellent effort", "Needs improvement"]
    },
    {
        "name": "Bob77",
        "grades": [10, 11, 14, 13],
        "comments": ["Solid progress", "Participates actively"]
    },
    {
        "name": "Charlie_K",
        "grades": [8, 7, 9, 10],
        "comments": ["Needs improvement", "Tries hard"]
    },
    {
        "name": "Diana202",
        "grades": [15, 14, 13, 16],
        "comments": ["Excellent student", "Very consistent"]
    },
    {
        "name": "Egor45",
        "grades": [9, 10, 8, 11],
        "comments": ["Can do better", "Pays attention in class"]
    }
]


def analyze_students(data):
    clean_students=[]

    for s in data:
        name=s["name"]

        if any(ch.isdigit() for ch in name):
            continue

    name=name.title()

    clean_students.append({
        "name":name,
        "grades":s["grades"],
        "comments":s["comments"]
                          })

    processed=[]
    for s in clean_students:
        for g in s["grades"]:
            if g<=0:
                continue
            if g<10 and g%2==1:
                g+=sum(int(d) for d in str(g))
            if g>=10 and g%2==0:
                g**=2
            processed.append(g)

        s["processed_grades"]=processed

    vowels=set("aeiou")
    all_vowels=set()
    student_words={}

    for s in clean_students:
        comments=s["comments"]

        text=" ".join(comments)

        for p in string.punctuation:
            text.replace(p,"")

        words=text.lower().split()

        unique_words=set()

        for w in words:
            if len(w)<4 or w==w[::-1]:
                continue
            else:
                unique_words.add(w)


            for ch in w:
                if ch in vowels:
                    all_vowels.add(ch)

        student_words[s["name"]]=unique_words


    word_counts={}

    for words in student_words.values():
        for word in words:
            if word not in word_counts:
                word_counts[word]=1
            else:
                word_counts[word]+=1



    word_counts={k:v for k,v in word_counts.items() if v>=2}

    word_counts=dict(sorted(word_counts.items(),key=lambda x:(-x[1],x[0])))


    students_by_avg=sorted(clean_students,key=lambda x:
    (-sum(x["processed_grades"])/len(x["processed_grades"]),x["name"]))

    students_by_length={}


    for s in clean_students:
        length=len(s["name"])

        if length not in students_by_length:
            students_by_length[length]=[]
        if s["name"] not in students_by_length[length]:
            students_by_length[length].append(s["name"])


    return {
        "students":clean_students,
        "word_counts":word_counts,
        "all_vowels":all_vowels,
        "students_by_avg":students_by_avg,
        "students_by_name_length":students_by_length
    }


print(analyze_students(students))


















```

```python
#Анализ заказов онлайн-магазина

data=[
    {
        "order_id": "ORD001",
        "customer": "alice",
        "items": [
            {"name": "Laptop", "price": 1200, "quantity": 1},
            {"name": "Mouse", "price": 25, "quantity": 2}
        ],
        "notes": [
            "Fast delivery please!",
            "Gift wrap if possible"
        ]
    },

    {
        "order_id": "ORD002",
        "customer": "bob",
        "items": [
            {"name": "Keyboard", "price": 80, "quantity": 1},
            {"name": "Monitor", "price": 300, "quantity": 2}
        ],
        "notes": [
            "Leave package at door.",
            "Call before arrival"
        ]
    },

    {
        "order_id": "ORD003",
        "customer": "charlie",
        "items": [
            {"name": "Headphones", "price": 150, "quantity": 1},
            {"name": "USB Cable", "price": 10, "quantity": 3}
        ],
        "notes": [
            "Fast shipping needed",
            "Thanks!"
        ]
    },

    {
        "order_id": "ORD004",
        "customer": "david",
        "items": [
            {"name": "Laptop", "price": 1100, "quantity": 1},
            {"name": "Mouse", "price": 25, "quantity": 1}
        ],
        "notes": [
            "Please pack carefully.",
            "Fragile item"
        ]
    },

    {
        "order_id": "ORD005",
        "customer": "eva",
        "items": [
            {"name": "Monitor", "price": 350, "quantity": 1},
            {"name": "Keyboard", "price": 90, "quantity": 1}
        ],
        "notes": [
            "Office use",
            "Need invoice"
        ]
    }
]

import string

def analyze_orders(orders):
    clean_orders=[]

    for o in orders:
        if any(ch.isdigit() for ch in o["customer"]):
            continue

        o["customer"]=o["customer"].title()

        clean_orders.append(o)


    for o in clean_orders:
        processed=[]

        for item in o["items"]:
            name=item["name"]
            price=item["price"]
            quantity=item["quantity"]

            if price<=0:
                continue

            if price>100 and quantity>1:
                price=price*quantity
            if quantity%2==1:
                price+=sum(int(d) for d in str(int(price)))

            processed.append({
                "name":name,
                "price":price,
                "quantity":quantity
            })
        o["processed_items"]=processed


    order_words={}
    all_vowels=set()
    vowels=set("aeiou")

    for o in clean_orders:
        text=" ".join(o["notes"])

        for p in string.punctuation:
            text=text.replace(p,"")

        words=text.lower().split()

        unique=set()

        for w in words:
            if len(w)<4:
                continue
            if w==w[::-1]:
                continue

            unique.add(w)

            for ch in w:
                if ch in vowels:
                    all_vowels.add(ch)


        order_words[o["order_id"]]=unique

    word_counts={}

    for words in order_words.values():
        for w in words:
            if w not in word_counts:
                word_counts[w]=1
            else:
                word_counts[w]+=1

    word_counts={k:v for k,v in word_counts.items() if v>=2}

    word_counts=dict(sorted(word_counts.items(),key=lambda x:(-x[1], x[0])))

    unique_products=set()

    for o in clean_orders:
        for item in o["processed_items"]:
            unique_products.add(item["name"])


    orders_by_total=sorted(clean_orders,key=lambda o:
    (-sum
    (i["price"] for i in o["processed_items"]),o["order_id"]))


    orders_by_item_count={}

    for o in clean_orders:
        c=len(o["processed_items"])

        if c not in orders_by_item_count:
            orders_by_item_count[c]=[o["order_id"]]
        else:
            orders_by_item_count[c].append(o["order_id"])


    return {
        "orders":clean_orders,
        "word_counts":word_counts,
        "all_vowels":all_vowels,
        "unique_products":unique_products,
        "orders_by_total":orders_by_total,
        "orders_by_item_count":orders_by_item_count
    }



analyze_orders(data)



```
