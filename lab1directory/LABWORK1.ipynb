```python
#7
import string
def palindrome_words(text):
    unique_palindroms=set()
    exclude=string.punctuation+"«»—"
    for p in exclude:
        text=text.replace(p,"")

    words=text.split()

    for word in words:
        if word==word[::-1] and len(word)>=3:
            unique_palindroms.add(word)

    unique_palindroms=sorted(unique_palindroms,key=lambda x:(-len(x),x))

    return unique_palindroms

t="Сегодня я гулял по парку, и, словно шутка судьбы, встретил маму, которая несла на руках уровень воды. Казалось, все вокруг сияло: солнце отражалось в озере, а лодка тихо скользила по воде. Вдруг я услышал, как кто-то сказал: А роза упала на лапу Азора! — и это было так странно, что я улыбнулся. Я подумал про себя: О, как чудесно видеть таких людей! Вечером, возвращаясь домой, я заметил на тротуаре слова «шалаш» и «казак», аккуратно выложенные детьми, которые играли неподалеку. Все это напомнило мне о том, что жизнь полна маленьких чудес, даже если иногда она кажется хаотичной."

print(palindrome_words(t))
```

```python
#19
def filter_by_digit_sum(nums):
    result = set()
    for num in nums:
        if num % 2 != 0:
            digit_sum = 0
            for digit in str(abs(num)):
                digit_sum += int(digit)
            if digit_sum % 2 == 0:
                result.add(num)
    return result

filter_by_digit_sum({12, 13, 21, 35, 44})
```

```python
#23
def group_by_last_letter(words):
    finaldict={}
    for a in words:
        if not a:
            continue
        b=a[-1]
        if b not in finaldict:
            finaldict[b]=[]
        if a not in finaldict[b]:
            finaldict[b].append(a)
    return finaldict

group_by_last_letter(["apple", "table", "banana", "grape", "apple", "orange", "pineapple"])
```

```python
#30
def partition_by_sum_parity(s):
    even_set=set()
    odd_set=set()
    for i in s:
        digit=[int(d) for d in str(i)]
        summa=sum(digit)
        if summa%2==0:
            even_set.add(i)
        else:
            odd_set.add(i)

    return (even_set,odd_set)

data={1, 5, 7, 12, 42, 100, 3, 8, 15, 9}

print(partition_by_sum_parity(data))

```

```python
#32
def pairwise_intersections(sets_list):
    if len(sets_list)<2:
        return []
    else:
        return [sets_list[i] & sets_list[i + 1] for i in range(len(sets_list) - 1)]

print(pairwise_intersections([{1, 2, 3}, {2, 3, 4}, {3, 4, 5}]))
```

```python
#38
from collections import Counter
def multi_symmetric_difference(sets_list):
    un=[]
    fin=[]
    for b in sets_list:
        for c in b:
            un+=[c]

    d=dict(Counter(un))
    for x,y in d.items():
        if y>1:
            continue
        fin+=[x]

    return f"симметрическая разность={set(fin)}"

multi_symmetric_difference([{3,4,5,6,7,8},{1,2,3,4,5}])
```

```python
#Анализ текстовой базы данных студентов

import json

import string

data=[
    {
        "name": "alice123",
        "grades": [12, 9, 15, 8],
        "comments": ["Good work!", "Needs practice.", "Good effort!"]
    },
    {
        "name": "bob77",
        "grades": [10, 11, 14, 13],
        "comments": ["Good work!", "Participates well."]
    },
    {
        "name": "charlieX",
        "grades": [8, 7, 9, 10],
        "comments": ["Needs practice.", "Should review lessons!"]
    },
    {
        "name": "diana202",
        "grades": [15, 14, 13, 15],
        "comments": ["Excellent effort!", "Good work!"]
    },
    {
        "name": "evan5",
        "grades": [6, 8, 7, 9],
        "comments": ["Needs practice.", "Good effort."]
    },
    {
        "name": "fionaZ",
        "grades": [13, 12, 14, 11],
        "comments": ["Good work!", "Clear answers."]
    },
    {
        "name": "george21",
        "grades": [9, 10, 8, 7],
        "comments": ["Needs practice.", "Slow progress..."]
    },
    {
        "name": "max",
        "grades": [11, 12, 10, 9],
        "comments": ["Good effort!", "Asks questions?"]
    },
    {
        "name": "leo99",
        "grades": [14, 13, 15, 14],
        "comments": ["Excellent effort!", "Good work!"]
    },
    {
        "name": "tom",
        "grades": [7, 6, 8, 7],
        "comments": ["Needs practice.", "Should review lessons!"]
    }
]

def analyze_students(students):
    clean_students=[]
    for s in students:
        name=s["name"]

        if any(ch.isdigit() for ch in name):
            continue

        clean_students.append({
            "name":name.title(),
            "grades":s["grades"],
            "comments":s["comments"]
        })


    for s in clean_students:
        processed_grades=[]

        grades=s["grades"]

        for grade in grades:
            if grade<=0:
                continue
            if grade<10 and grade%2==1:
                grade+=sum(int(d) for d in str(grade))
                processed_grades.append(grade)
            if grade>=10 and grade%2==0:
                grade**=2
                processed_grades.append(grade)
            else:
                processed_grades.append(grade)

        s["grades"]=processed_grades

    unique_words=set()
    unique_words_list=[]
    all_vowels=set()
    vowels="aeiou"

    for s in clean_students:
        comments=s["comments"]

        text=" ".join(comments)

        for p in string.punctuation:
            text=text.replace(p,"")

        splitted_text=text.split()

        for word in splitted_text:
            if len(word)>=4 and word!=word[::-1]:
                unique_words.add(word)

        for word in unique_words:
            if word not in unique_words_list:
                unique_words_list.append(word)
            for ch in word:
                if ch in vowels:
                    all_vowels.add(ch)

    word_dictionary={}

    for s in clean_students:
        comments=s["comments"]

        for comment in comments:
            if comment not in word_dictionary:
                word_dictionary[comment]=1
            else:
                word_dictionary[comment]+=1


    word_dictionary={k:v for k,v in word_dictionary.items() if v>=2}

    word_dictionary=dict(sorted(
        word_dictionary.items(),key=lambda x:
        x[0]
    ))

    students_by_avg=sorted(
        clean_students,key=lambda x:
        (-sum(x["grades"])/len(x["grades"]),x["name"])
    )

    students_by_avg=[s["name"] for s in students_by_avg]


    students_by_length={}

    for s in clean_students:
        name=s["name"]

        length=len(name)

        if length not in students_by_length:
            students_by_length[length]=[]
        if name not in students_by_length[length]:
            students_by_length[length].append(name)

    return {
        "students with processed name and grades":clean_students,
        "word count":word_dictionary,
        "all_vowels":list(all_vowels),
        "students_by_avg":students_by_avg,
        "students_by_length":students_by_length
    }




result_students=analyze_students(data)

with open("lab1students.json","w",encoding="utf-8")as f:
    json.dump(result_students,f,indent=4,ensure_ascii=False)






```

```python
#Анализ заказов онлайн-магазина

import json

import string

data=[
    {
        "order_id": "A101",
        "customer": "alice",
        "items": [
            {"name": "Laptop", "price": 899.99, "quantity": 1},
            {"name": "Mouse", "price": 20.5, "quantity": 2},
            {"name": "Keyboard", "price": 45.0, "quantity": 1}
        ],
        "notes": ["Deliver after 5pm!", "Fragile package.", "Pack carefully."]
    },
    {
        "order_id": "A102",
        "customer": "bob77",
        "items": [
            {"name": "Monitor", "price": 199.99, "quantity": 2},
            {"name": "HDMI cable", "price": 12.99, "quantity": 3}
        ],
        "notes": ["Call before delivery.", "Fragile package."]
    },
    {
        "order_id": "A103",
        "customer": "charlie",
        "items": [
            {"name": "Desk lamp", "price": 35.5, "quantity": 1},
            {"name": "Notebook", "price": 5.25, "quantity": 5},
            {"name": "Pen set", "price": 9.99, "quantity": 2}
        ],
        "notes": ["Gift wrap, please.", "Include receipt!", "Pack carefully."]
    },
    {
        "order_id": "A104",
        "customer": "diana21",
        "items": [
            {"name": "Backpack", "price": 49.99, "quantity": 1},
            {"name": "Water bottle", "price": 15.0, "quantity": 2}
        ],
        "notes": ["Leave at the door.", "Call before delivery."]
    },
    {
        "order_id": "A105",
        "customer": "eric",
        "items": [
            {"name": "Tablet", "price": 320.0, "quantity": 1},
            {"name": "Stylus", "price": 18.75, "quantity": 1},
            {"name": "Screen protector", "price": 7.5, "quantity": 2}
        ],
        "notes": ["Fragile package!", "Priority shipping.", "Pack carefully."]
    }
]

def analyze_orders(orders):
    clean_orders=[]
    for o in orders:
        customer_name=o["customer"]

        if any(ch.isdigit() for ch in customer_name):
            continue

        clean_orders.append({
            "order_id":o["order_id"],
            "customer":customer_name.title(),
            "items":o["items"],
            "notes":o["notes"]
        })

    processed_items=[]
    for o in clean_orders:
        items=o["items"]

        for item in items:
            name=item["name"]
            price=item["price"]
            quantity=int(item["quantity"])

            if price<=0:
                continue
            if price>100 and quantity>1:
                price=price*quantity
            if quantity%2==1:
                price+=sum(int(d) for d in str(int(price)))

            processed_items.append({
                "name":name,
                "price":price,
                "quantity":quantity
            })

        o["items"]=processed_items


    unique_words=set()
    vowels="aeiou"
    all_vowels=set()

    for o in clean_orders:
        notes=o["notes"]

        text=" ".join(notes)

        for p in string.punctuation:
            text=text.replace(p,"")


        splitted=text.split()

        for word in splitted:
            if len(word)>=4 and word!=word[::-1]:
                unique_words.add(word)

        for word in unique_words:
            for ch in word:
                if ch in vowels:
                    all_vowels.add(ch)

    unique_words=list(unique_words)

    word_dictionary={}

    for o in clean_orders:
        notes=o["notes"]

        for note in notes:
            if note not in word_dictionary:
                word_dictionary[note]=1
            else:
                word_dictionary[note]+=1

    word_dictionary={k:v for k,v in word_dictionary.items() if v>=2}

    word_dictionary=dict(sorted(
        word_dictionary.items(),
        key=lambda x:(-x[1],x[0])
    ))

    unique_products=set()

    for o in clean_orders:
        items=o["items"]

        for item in items:
            name=item["name"]

            if name not in unique_products:
                unique_products.add(name)

    orders_by_total=sorted(
        clean_orders,
        key=lambda x:(-sum(y["price"] for y in x["items"]),x["order_id"])
    )

    orders_by_total=[o["order_id"] for o in orders_by_total]

    orders_by_item_count={}

    for o in clean_orders:
        order_id=o["order_id"]
        length=len(o["items"])

        if length not in orders_by_item_count:
            orders_by_item_count[length]=[]
        if order_id not in orders_by_item_count[length]:
            orders_by_item_count[length].append(order_id)

    return {
        "ORDERS WITH PROCESSED NAME AND ITEMS":clean_orders,
        "word count":word_dictionary,
        "vowels":list(all_vowels),
        "orders_by_total":orders_by_total,
        "orders_by_item":orders_by_item_count
    }




result=analyze_orders(data)

with open("lab1orders.json","w",encoding="utf-8")as f:
    json.dump(result,f,indent=4,ensure_ascii=False)



```
