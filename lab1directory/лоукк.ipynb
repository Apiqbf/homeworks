```python
#2
import csv
e=[["Ali","IT",500000],
["Dana","HR",300000],
["Arman","IT",600000],
["Aruzhan","Marketing",400000],
["Dias","IT",450000]
]

with open("csvfile.csv","w",newline="",encoding="utf-8")as c:
    writer=csv.writer(c)

    writer.writerow(["name","department","salary"])
    writer.writerows(e)
```

```python
#2
import csv

total=0
departments={}
employees=[]
count=0

with open("csvfile.csv","r",encoding="utf-8")as f:
    reader=csv.DictReader(f)

    for a in reader:
        name=a["name"]
        department=a["department"]
        salary=int(a["salary"])

        total+=salary
        count+=1

        employees.append({"department":department,"name":name,"salary":salary})

        if department not in departments:
            departments[department]={"salary":salary,"count":1}
        else:
            departments[department]["salary"]+=salary
            departments[department]["count"]+=1


avg=total/count

dep_avg={}
for dep in departments:
    dep_avg[dep]=departments[dep]["salary"]/departments[dep]["count"]

highest=max(employees,key=lambda x:x["salary"])

higher_than_avg=[emp for emp in employees if emp["salary"]>avg]


with open("high_sal.csv","w",newline="",encoding="utf-8")as report:
    field=["department","name","salary"]
    writer=csv.DictWriter(report,fieldnames=field)
    writer.writeheader()
    writer.writerows(higher_than_avg)


with open("report.loukk","w",encoding="utf-8")as r:
    r.write(f"Средняя зарплата:{avg}\n\n")
    r.write(f"Cредняя зарплата по отделам:\n")
    for dep,sal in dep_avg.items():
        r.write(f"{dep}:{sal}\n")

    r.write(f"\nОтдел с самой высокой средней зарплатой:{max(dep_avg,key=dep_avg.get)}\n")
    r.write(f"Самый высокооплачиваемый сотрудник:{highest.get("name")}")
```

```python
#4
import csv
a=[["user_1",5000],
["user_2",10000],
["user_1",700000],
["user_3",3000],
["user_2",900000],
["user_4",2000]
]
with open("trans.csv","w",newline="",encoding="utf-8")as c:
    writer=csv.writer(c)
    writer.writerow(["user_id","amount"])
    writer.writerows(a)
```

```python
import csv
import json

user_frequency={}
s_transactions=[]
s_users=set()
summ=0

with open("trans.csv","r",encoding="utf-8")as file:
    reader=csv.DictReader(file)
    for a in reader:
        user=a["user_id"]
        amount=int(a["amount"])

        if user not in user_frequency:
            user_frequency[user]=1
        else:
            user_frequency[user]+=1

        if amount>5000:
            s_transactions+=[{"user_id":user,"amount":amount}]
            summ+=1
            s_users.add(user)




for user,count in user_frequency.items():
    if count>3:
        s_users.add(user)

with open("f_report.txt","w",encoding="utf-8")as r:
    r.write(f"Подозрительных транзакции:{len(s_transactions)}\n")
    r.write(f"Подозрительных пользователей:{len(s_users)}\n")
    r.write(f"Пользователи:\n")
    for b in s_users:
        r.write(f"{b}\n")
    r.write(f"\nОбщая сумма подозрительных транзакции:{summ}\n")

with open("f.json","w",encoding="utf-8")as f:
    json.dump(list(s_users),f,indent=4,ensure_ascii=False)
```

```python
import json

total=0
order_per_user={}
top_us=""
max_ord=0
item_freq={}
most_popular=("")
max_count=0

with open("o.json","r",encoding="utf-8")as f:
    reader=json.load(f)

    for a in reader:
        user=a["user"]
        items=a["items"]
        amount=int(a["total"])

        total+=amount

        if amount>max_ord:
            max_ord=amount
            top_us=user

        if user not in order_per_user:
            order_per_user[user]=1
        else:
            order_per_user[user]+=1

        for item in items:
            if item not in item_freq:
                item_freq[item]=1
            else:
                item_freq[item]+=1

for i in item_freq:
    if item_freq[i]>max_count:
        most_popular=item
        max_count=item_freq[i]


itog={
    "total_revenue":total,
    "top_user":top_us,
    "most_popular":most_popular,
    "total orders":len(reader)
}


with open("summary.json","w",encoding="utf-8")as r:
    json.dump(itog,r,indent=4,ensure_ascii=False)
```
