```python
with open("txt,csv files/data.txt", "w", encoding="utf-8")as f:
    f.write("hello world\n")
with open("txt,csv files/data.txt", "r", encoding="utf-8")as f:
    a=f.read()
    print(a)
```

```python
a=[12,436,345,22]
with open("txt,csv files/data.txt", "a", encoding="utf-8")as f:
    for b in a:
        f.write(str(b)+"\n")
```

```python
a=["asem","dimash","samgar"]
with open("txt,csv files/names.txt", "w", encoding="utf-8")as f:
    for b in a:
        f.write(b.capitalize()+"\n")

with open("txt,csv files/names.txt", "r", encoding="utf-8")as f:
    r=f.read()
    print(r)
```
