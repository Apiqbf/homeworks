```python
#1
users=set()
total_buys=0
summa=0
top_user=""
user_spending={}

file=open("task1/shop_logs.txt","r")

for a in file:
    parts=a.strip().split(";")
    user=parts[1]
    action=parts[2]

    users.add(user)

    if action=="BUY":
        total_buys+=1
        amount=int(parts[3])
        summa+=amount

    if user not in user_spending:
        user_spending[user]=1
    else:
        user_spending[user]+=1

max=0
for u in user_spending:
    if user_spending[u]>max:
        max=user_spending[u]
        top_user=u

avg=summa/total_buys
print(user_spending)
with open("task1/report.txt", "w", encoding="utf-8")as r:
    r.write(f"Уникальных пользователей:{users}\n")
    r.write(f"Всего покупок:{total_buys}\n")
    r.write(f"Общая сумма-{summa}\n")
    r.write(f"Самый активный пользователь:{top_user}\n")
    r.write(f"Средний чек:{avg}")




```

```python
#2
import csv

employees = [
    ["Ali","IT",500000],
    ["Dana","HR",300000],
    ["Arman","IT",600000],
    ["Aruzhan","Marketing",400000],
    ["Dias","IT",450000],
]

with open("task2/employees.csv", "w", newline="", encoding="utf-8")as file:
    writer=csv.writer(file)

    writer.writerow(["name","department","salary"])
    writer.writerows(employees)
```

```python
#2
import csv

total_salary = 0
count = 0
department_data = {}
employees = []

with open("task2/employees.csv", "r", encoding="utf-8") as file:
    reader = csv.DictReader(file)

    for row in reader:
        name = row["name"]
        department = row["department"]
        salary = int(row["salary"])

        employees.append({"name": name, "department": department, "salary": salary})

        total_salary += salary
        count += 1

        if department not in department_data:
            department_data[department] = {"sum": salary, "count": 1}
        else:
            department_data[department]["sum"] += salary
            department_data[department]["count"] += 1

average_salary = total_salary / count

department_average = {}
for dep in department_data:
    department_average[dep] = department_data[dep]["sum"] / department_data[dep]["count"]

top_department = max(department_average, key=department_average.get)

top_employee = max(employees, key=lambda x: x["salary"])

high_salary_employees = [emp for emp in employees if emp["salary"] > average_salary]

with open("task2/high_salary.csv", "w", newline="", encoding="utf-8") as file:
    fieldnames = ["name", "department", "salary"]
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(high_salary_employees)

with open("task2/salary_report.txt", "w", encoding="utf-8") as report:
    report.write(f"Средняя зарплата: {average_salary}\n\n")

    report.write("Средняя зарплата по отделам:\n")
    for dep, avg in department_average.items():
        report.write(f"{dep}: {avg}\n")

    report.write(f"\nОтдел с самой высокой средней зарплатой: {top_department}\n")
    report.write(f"Самый высокооплачиваемый сотрудник: {top_employee['name']}\n")
```

```python
#3
import json

data=[
  {
    "order_id": 1,
    "user": "Ali",
    "items": ["phone", "case"],
    "total": 300000
  },
  {
    "order_id": 2,
    "user": "Dana",
    "items": ["laptop"],
    "total": 800000
  },
  {
    "order_id": 3,
    "user": "Ali",
    "items": ["mouse", "keyboard"],
    "total": 70000
  }
]

with open("task3/orders.json", "w", encoding="utf-8")as f:
    json.dump(data,f,indent=4,ensure_ascii=False)

```

```python
summa=0
top_user=""
most_popular_item=""
max_item=0
max_spent=0
item_frequency={}
count=0

with open("task3/orders.json", "r", encoding="utf-8")as f:
    reader=json.load(f)

    for a in reader:
        user=a["user"]
        items=a["items"]
        total=int(a["total"])

        if total>max_spent:
            max_spent=total
            top_user=user

        summa+=total

        for i in items:
            count+=1
            if i not in item_frequency:
                item_frequency[i]=1
            else:
                item_frequency[i]+=1

for i in item_frequency:
    if item_frequency[i]>max_item:
        max_item=item_frequency[i]
        most_popular_item=i

summary={
    "total_revenue":summa,
    "top_user":top_user,
    "most_popular_item":most_popular_item,
    "total_orders":count
}

with open("task3/summary.json", "w", encoding="utf-8")as j:
    json.dump(summary,j,indent=4,ensure_ascii=False)
```

```python
#4
import csv

a=[["user_1",5000],
["user_2",10000],
["user_1",700000],
["user_3",3000],
["user_2",900000],
["user_4",2000]
]
with open("task4/transactions.csv", "w", newline="", encoding="utf-8")as f:
    writer=csv.writer(f)
    writer.writerow(["user_id","amount"])
    writer.writerows(a)
```

```python
#4
import csv
import json

sus_transactions=[]
user_operation_count={}
sus_users=set()
sus_transactions_sum=0

with open("task4/transactions.csv", "r", encoding="utf-8")as f:
    reader=csv.DictReader(f)
    for a in reader:
        user=a["user_id"]
        amount=int(a["amount"])

        if user not in user_operation_count:
            user_operation_count[user]=1
        else:
            user_operation_count[user]+=1

        if amount>5000:
            sus_transactions.append({"user_id":user,"amount":amount})
            sus_transactions_sum+=1
            sus_users.add(user)


for user,count in user_operation_count.items():
    if count>3:
        sus_users.add(user)

with open("task4/fraud_report.txt", "w", encoding="utf-8")as txt:
    txt.write(f"Подозрительынх транзакции:{len(sus_transactions)}\n")
    txt.write(f"Подозрительных пользователей:{len(sus_users)}\n")
    txt.write(f"Список пользователей:{sus_users}\n")
    txt.write(f"Общая сумма подозрительных транзакции:{sus_transactions_sum}")


with open("task4/fraud_users.json", "w", encoding="utf-8")as js:
    json.dump(list(sus_users),js,indent=4,ensure_ascii=False)
```
