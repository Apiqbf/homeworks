```python
#22
def group_by_parity_and_sort(nums):
    even=[]
    odd=[]
    for a in nums:
        if a%2==0:
            even+=[a]
        else:
            odd+=[a]
    return even+odd

group_by_parity_and_sort([1,2,657,5,6,4,18])

```

```python
#26
def remove_duplicates_keep_last(nums):
    result=[]
    for a in reversed(nums):
        if a not in result:
            result+=[a]
    result.reverse()
    return result

remove_duplicates_keep_last([1, 2, 3, 2, 4, 1, 5])
```

```python
#11
def group_by_length(words):
    result={}
    for w in words:
        l=len(w)
        if l not in result:
            result[l]=[]
        if w not in result:
            result[l]+=[w]
    return result

group_by_length(["lkflkd","dlsfn","sdf","sd","sgg","f","f","a","fwef"])
```

```python
#24
def union_of_filtered_sets(sets_list):
    new=[]
    for b in sets_list:
        for c in b:
            if c<10 or c%2==0:
                continue
            new+=[c]
    print(set(new))

union_of_filtered_sets([{34,3,56,7,8,9,15},{5,6,7,89},{13,3444,50}])
```

```python
#36
from itertools import combinations
def all_subsets_of_size_k(s,k):
    new=[]
    for b in combinations(s,k):
        new+=[set(b)]
    return new


all_subsets_of_size_k({1,2,3,4,5},4)

```

```python
#38
from collections import Counter
def multi_symmetric_difference(sets_list):
    un=[]
    fin=[]
    for b in sets_list:
        for c in b:
            un+=[c]

    d=dict(Counter(un))
    for x,y in d.items():
        if y>1:
            continue
        fin+=[x]

    return f"симметрическая разность={set(fin)}"

multi_symmetric_difference([{3,4,5,6,7,8},{1,2,3,4,5}])
```
