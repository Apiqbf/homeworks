```python
#LAMBDA функции
#1
check=lambda a:"положительное" if a>0 else "отрицательное" if a!=0 else "ноль"

print(check(0))
print(check(5))
print(check(-5))
```

```python
#2
words = ["арбуз", "кот", "машина", "дом", "ананас"]

words=sorted(words,key=lambda x:
(len(x),x)
             )

print(words)
```

```python
#3
numbers = [5, 12, 7, 20, 33, 8]

result=list(filter(lambda x:x%2==0 and x>10,numbers))

print(result)
```

```python
#4
numbers = [1, 2, 3, 4, 5, 6]

result=list(map(lambda x:x**2 if x%2==0 else x*3,numbers))

print(result)
```

```python
#5
f=lambda a,b:"a больше" if a>b else "b больше" if a!=b else "равны"

print(f(7,7))
print(f(3,4))
print(f(4,3))
```

```python
#6
numbers = [0, -3, 5, -7, 8]

new=list(map(lambda x:"положительное" if x>0 else "отрицательное" if x!=0 else "ноль"
             ,numbers))

print(new)
```

```python
#1
def even_numbers(n):
    i=0
    while i<=n:
        yield i
        i+=2

        if i==4:
            yield "кратно 4"

for x in even_numbers(10):
    print(x)
```

```python
#1
numbers=[n**2 for n in range(1,21) if n%2==0]

print(numbers)
```

```python
#3
words=["кот","машина","ананас","дом"]

words=[w for w in words if len(w)>4
       and all(ch!="а" for ch in w)]

print(words)
```

```python
#4
numbers = [1,2,3,4,5]

numbers_dict={n:"четное" if n%2==0 else "нечетное" for n in numbers}

print(numbers_dict)
```

```python
#6
new=["FizzBuzz" if n%3==0 and n%5==0
     else "Fizz" if n%3==0
     else "Buzz" if n%5==0
     else n
     for n in range(1,21)]


print(new)
```
